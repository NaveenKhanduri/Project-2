```python
file_to_parse_aloan = "LoanStats3a_securev1.csv"
file_to_parse_rloan = "RejectStatsA.csv"
#file_to_parse_population = "../DataGun/population.csv"

df_aloan = pd.DataFrame(pd.read_csv(file_to_parse_aloan))
df_rloan = pd.DataFrame(pd.read_csv(file_to_parse_rloan))
df_rloan.head(5)
```

```python
print(df_aloan.columns.values)
df_aloan.columns.values[1]
```

```python
df_aloan["fico_range_avg"] = (df_aloan["fico_range_low"] + df_aloan["fico_range_high"])/2.0
df_aloan.head()
```

```python
# Data Cleanup
df_sel_aloan = df_aloan[["loan_amnt","fico_range_avg","dti","emp_length","policy_code"]]
df_rloan = df_rloan.rename(columns = {"Amount Requested":"loan_amnt","Policy Code":"policy_code" ,"Risk_Score":"fico_range_avg","Debt-To-Income Ratio":"dti","Employment Length":"emp_length"})
df_sel_rloan = df_rloan[["loan_amnt","fico_range_avg","dti","emp_length","policy_code"]]
df_sel_rloan = df_sel_rloan.iloc[:50000]
#df_sel_rloan = df_aloan[["Amount Requested","Risk_Score","Debt-To-Income Ratio","Employment Length"]]

#df_sel_aloan.head()
#df_sel_rloan.head()

#df_sel_aloan=df_sel_aloan.iloc[0:42538,:]
```

```python
#df_combine['emp_length'].replace(['<','+','years','year'],'')
df = pd.concat([df_sel_aloan,df_sel_rloan])
df['emp_length'].replace(['<','+','years','year'],'')
df.replace(['<','+','years','year'],'')

for toRep in ['<','+','years','year']:    
   df['emp_length']=df['emp_length'].str.replace(toRep,'') 

df.emp_length=pd.to_numeric(df.emp_length)
df.to_csv('combine_data.csv')
df.head()
```
